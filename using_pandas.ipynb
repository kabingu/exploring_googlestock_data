HOW TO EXPLORE WEB DATA USING API'S AND PANDAS


```python
#import libraries 
from alpha_vantage.timeseries import TimeSeries
import requests
import pandas as pd
from bs4 import BeautifulSoup
import io
```

```python
#open text file with alpha vantage API key
with open("api_key.txt") as file:
    api_key = file.read()
api_key = api_key.strip()
```

```python
# get weekly stock data of google in DatFrame format
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY&symbol=GOOGL&apikey=' + str(api_key) + "&datatype=csv"
```

```python
r = requests.get(url).content
data = pd.read_csv(io.StringIO(r.decode("utf-8")))
print(data)



```

PYTHON PACKAGE METHOD


```python
#store google data intraday data in data and meta_data variables 

ts1 = TimeSeries(key = api_key)
google_1, meta_data = ts1.get_intraday('GOOGL')
```

```python
#print metadata
meta_data
```

```python
#store data in a dataframe
df_google1 = pd.DataFrame(google_1).transpose().reset_index()

df_google1.head(10)
```

```python
#pandas dataframe automatically 
ts2 = TimeSeries(key=api_key, output_format="pandas")
ts2.get_intraday("GOOGL")
```

```python
#separate into metadata and intraday
df_google2, meta_data2 = ts2.get_intraday('GOOGL', outputsize="full")
meta_data2
```
